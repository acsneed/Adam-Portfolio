```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Part 1: Loading the Dataset; Preliminary Analysis

In this part, we will load the data from diamonds.txt into a DataFrame, then explore its structure.

```python
diamonds = pd.read_csv("diamonds.txt", delimiter="\t")
print(diamonds.head(10))
```

We will now determine the size of the dataset.

```python
print(diamonds.shape)
```

We will now inspect the distribution of columns in the dataset.

```python
display(diamonds.describe())
```

## Part 2: Filtering and Sorting

We will start this part by viewing information about the 5 most expensive diamonds in the dataset.

```python
result = diamonds[['price', 'carat', 'cut', 'color', 'clarity']] \
            .sort_values(by='price', ascending=False) \
            .head(5)

display(result)
```

We will now view information about the 5 least expensive diamonds in the dataset.

```python
result2 = diamonds[['price', 'carat', 'cut', 'color', 'clarity']] \
            .sort_values(by='price', ascending=True) \
            .head(5)

display(result2)
```

We will now view information about the 5 largest diamonds in the dataset with an
ideal cut.

```python
result3 = diamonds[['price', 'carat', 'cut', 'color', 'clarity']] \
            [diamonds['cut'] == 'Ideal'] \
            .sort_values(by='carat', ascending=False) \
            .head(5)

display(result3)
```

We will now view information about the 5 largest diamonds in the dataset with an
fair cut

```python
result4 = diamonds[['price', 'carat', 'cut', 'color', 'clarity']] \
            [diamonds['cut'] == 'Fair'] \
            .sort_values(by='carat', ascending=False) \
            .head(5)

display(result4)
```

## Part 3: Working with Categorical Variables

In this part, we will be creating lists to specify the order for cut, color, and clarity.

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

We will now use these lists to communicate to pandas the correct order of the levels.

```python
diamonds.cut = pd.Categorical(diamonds.cut, cut_levels)
diamonds.color = pd.Categorical(diamonds.color, color_levels)
diamonds.clarity = pd.Categorical(diamonds.clarity, clarity_levels)
```

We will now create lists of named colors that we will used for visualizations later in the project.

```python
clarity_pal = ['red', 'blue', 'green', 'magenta', 'darkgreen', 'yellow', 'orange', 'cyan']
color_pal = ['red', 'blue', 'green', 'magenta', 'orange']
cut_pal = ['red', 'blue', 'green', 'magenta', 'yellow', 'darkgreen', 'orange']
```

## Part 4: Displaying Counts for Categorical Variables

We will start this part by counting the number of diamonds for each level of cut.

```python
display(diamonds.cut.value_counts().sort_index())
```

We will now count the number of diamonds for each level of color.

```python
display(diamonds.color.value_counts().sort_index())
```

We will now count the number of diamonds for each level of clarity.

```python
display(diamonds.clarity.value_counts().sort_index())
```

## Part 5: Scatterplots of Price Against Carat

In this part, we will create scatter plots to find the relationship between price and carat.

```python
plt.figure(figsize=[8, 6])

plt.scatter(diamonds.carat, diamonds.price, s=20, alpha=0.4, c='blue', edgecolors='none')

plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')

plt.show()
```

Next, we will create the same plot, but will now color the points according to their clarity.

```python
plt.figure(figsize=[8, 6])

clarity_levels = diamonds.clarity.unique().tolist()
price = diamonds.price.tolist()
carat = diamonds.carat.tolist()

for i, level in enumerate(clarity_levels):
    plt.scatter(diamonds.carat[diamonds.clarity == level], diamonds.price[diamonds.clarity == level], s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none', label = level)

plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')

plt.legend()

plt.show()
```

We will now separate each clarity level into its own subplot.

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
fig.suptitle('Diamonds Scatter Plots by Clarity Levels')

for i, level in enumerate(clarity_levels):
    row = i // 4
    col = i % 4
    ax = axes[row, col]
    ax.scatter(diamonds.carat[diamonds.clarity == level], diamonds.price[diamonds.clarity == level], s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none')
    ax.set_xlim([0, 4])
    ax.set_ylim([0, 20000])
    ax.set_xlabel('Carat Size')
    ax.set_ylabel('Price (in $)')
    ax.set_title(level)

plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations

We will start this part by adding two new columns to the diamonds so we can store the transformed variables.

```python
diamonds['ln_carat'] = np.log(diamonds['carat'])
diamonds['ln_price'] = np.log(diamonds['price'])
```

We will now explore the distribution of the diamond prices, and the log of diamond prices using histograms.

```python
plt.figure(figsize=[12, 4])

plt.subplot(1, 2, 1)
plt.hist(diamonds['price'], bins=20, edgecolor='black', color='skyblue')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Prices')

plt.subplot(1, 2, 2)
plt.hist(diamonds['ln_price'], bins=20, edgecolor='black', color='lightgreen')
plt.xlabel('Natural Log of Price')
plt.ylabel('Count')
plt.title('Histogram of Log-Prices')

plt.tight_layout()

plt.show()
```

We will now use histograms to explore the distribution of the carat sizes, and the log of carat sizes.

```python
plt.figure(figsize=[12, 4])

plt.subplot(1, 2, 1)
plt.hist(diamonds['carat'], bins=20, edgecolor='black', color='lightblue')
plt.xlabel('Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Carat Sizes')

plt.subplot(1, 2, 2)
plt.hist(diamonds['ln_carat'], bins=20, edgecolor='black', color='lightgreen')
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Log-Carat Sizes')

plt.tight_layout()

plt.show()
```

## Part 7: Scatterplot of Transformed Variables

In this part, we will create a scatterplot of ln_price against ln_carat, with points colored according to their clarity.

```python
plt.figure(figsize=[8, 6])

clarity_levels = diamonds.clarity.unique().tolist()

ln_price = diamonds.ln_price.tolist()
ln_carat = diamonds.ln_carat.tolist()

for i, level in enumerate(clarity_levels):
    plt.scatter(diamonds.ln_carat[diamonds.clarity == level], diamonds.ln_price[diamonds.clarity == level], s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none', label = level)


plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title('Relationship between Log-Price and Log-Carat Size')

plt.legend()

plt.show()
```

## Part 8: Grouping by Cut

In this part, we will create a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
gb_cut = diamonds[['cut', 'price', 'carat']].groupby('cut').mean()
display(gb_cut)
```

We will now create bar charts to graphically display the information found above.

```python
plt.figure(figsize=[12, 4])

plt.subplot(1, 2, 1)
cut_pal = {'Fair': 'red', 'Good': 'orange', 'Very Good': 'green', 'Premium': 'blue', 'Ideal': 'purple'}
plt.bar(gb_cut.index, gb_cut['price'], edgecolor='black', color=[cut_pal[cut] for cut in gb_cut.index])
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')

plt.subplot(1, 2, 2)
plt.bar(gb_cut.index, gb_cut['carat'], edgecolor='black', color=[cut_pal[cut] for cut in gb_cut.index])
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Cut')

plt.tight_layout()

plt.show()
```
